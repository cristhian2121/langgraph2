```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain import chat_models
llm = chat_models.init_chat_model("openai:gpt-4o-mini", temperature=0)
```

```python
from pydantic import BaseModel, Field

class ContactInfo(BaseModel):
    name: str = Field(description="The name of the person")
    email: str = Field(description="The email of the person")
    phone: str = Field(description="The phone number of the person")
    tone: int = Field(description="The tone of the person", ge=0, le=100)
    age: int = Field(description="The age of the person")
    sentiment: str = Field(description="The sentiment of the person")
```

```python
llm_with_structured_output = llm.with_structured_output(ContactInfo)
```

```python
messages = [
    ("system", "You are a helpful assistant that can extract information from a given conversation. If the user doesn't provide the information, return None for that field."),
    
    ("user", "Hola Soy Pepe tengo 33 a√±os, voy a ser padre estoy abrumado")
]

response = llm_with_structured_output.invoke(messages)
print(response)
```
